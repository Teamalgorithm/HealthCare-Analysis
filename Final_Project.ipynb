```python
# to check the duplicates
ph.duplicated().sum()
```

```python
#to capitalize first letter of each word and remove extra space from beginning and end
ph.Name=ph.Name.str.strip().str.title()
```

```python
ph
```

```python
#to capitalize first letter, and remove extra space from beginning and end
ph.Gender=ph.Gender.str.strip().str.capitalize()
```

```python
ph
```

```python
# to find rows that are duplicated in all columns except age
dupes=ph[ph.duplicated(subset=['P_ID','Name','Gender','Blood Type'],keep=False)]
print(len(dupes))
```

```python
# to fill missing ages with the avg age of each gender
ph['Age']=ph['Age'].fillna(ph.groupby('Gender')['Age'].transform('mean'))
```

```python
# to check the nulls
ph.isnull().sum()
```

```python
# to change age column data type
ph.Age=ph.Age.astype(int)
```
